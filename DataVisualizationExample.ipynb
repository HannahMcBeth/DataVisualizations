```
import numpy as np  #numpy provides fast mathematical function processing
import pandas as pd #pandas is used for easy indexing and data collecting tools
import matplotlib as mapl #A library for high-level data visualization
import matplotlib.pyplot as plt
%matplotlib inline 
```

```
#my data is called govenor's county so the dataframe is named GovCounty
GovCounty=pd.read_csv("https://raw.githubusercontent.com/hsbaum/DataVisualizations/main/governors_county.csv")
#pd.read_csv is code to import csv files
```

```
#rename dataframe to not overwrite original
#execute code
#pick two columns
GovCounty2 = GovCounty.plot.line(x='county', y='percent')

plt.title('Line Plot')
plt.xlabel('County')
plt.ylabel('Percent Reporting')
plt.rcParams['figure.figsize'] = {50,20}
plt.show
```

```
#rename dataframe to not overwrite original
#execute code
#pick two columns
GovCounty3 = GovCounty.plot.bar(x='county', y='percent')

plt.title('Bar Plot')
plt.xlabel('County')
plt.ylabel('Percent Reporting')
plt.rcParams['figure.figsize'] = {50,20}
plt.show
```
